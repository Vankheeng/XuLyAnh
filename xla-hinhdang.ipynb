## Nhận diện hình dạng đơn giản bằng CNN

```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import seaborn as sns


# 1. LOAD VÀ XỬ LÝ DATASET
def load_and_prepare_data(dataset_path):
    """Load dataset"""
    
    images = []
    labels = []
    class_names = ['Circle', 'Square', 'Triangle']
    class_mapping = {name: idx for idx, name in enumerate(class_names)}
    
    # Kiểm tra cấu trúc dataset
    for item in os.listdir(dataset_path):
        item_path = os.path.join(dataset_path, item)
        if os.path.isdir(item_path):
            print(f" {item}: {len(os.listdir(item_path))} files")
    
    # Load 
    for class_name in class_names:
        class_path = os.path.join(dataset_path, class_name)
        
        if not os.path.exists(class_path):
            print(f"Thư mục {class_name} không tồn tại, bỏ qua...")
            continue
            
        print(f"Đang xử lý {class_name}...")
        count = 0
        
        for img_file in os.listdir(class_path):
            if img_file.lower().endswith(('.png', '.jpg', '.jpeg')):
                img_path = os.path.join(class_path, img_file)
                
                try:
                    # Đọc và tiền xử lý ảnh
                    img = cv2.imread(img_path)
                    if img is None:
                        continue
                        
                    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                    img = cv2.resize(img, (128, 128))
                    img = img.astype('float32') / 255.0
                    
                    images.append(img)
                    labels.append(class_mapping[class_name])
                    count += 1
                    
                except Exception as e:
                    print(f"Lỗi khi xử lý {img_path}: {e}")
                    continue
        
        print(f"{class_name}: {count} ảnh")
    
    images = np.array(images)
    labels = tf.keras.utils.to_categorical(labels, len(class_names))
    
    print(f"Tổng số ảnh: {len(images)}")
    print(f"Số lớp: {len(class_names)}")
    
    return images, labels, class_names

# 2. XÂY DỰNG CNN MODEL
class ShapeDetectorCNN:
    def __init__(self, input_shape=(128, 128, 3), num_classes=3):
        self.input_shape = input_shape
        self.num_classes = num_classes
        self.model = None
        self.class_names = ['circle', 'square', 'triangle']
        self.history = None
        
    def build_model(self):
        """Xây dựng CNN model cho 3 lớp"""
        model = keras.Sequential([
            # Block 1
            layers.Conv2D(32, (3, 3), activation='relu', input_shape=self.input_shape),
            layers.BatchNormalization(),
            layers.MaxPooling2D((2, 2)),
            layers.Dropout(0.25),
            
            # Block 2
            layers.Conv2D(64, (3, 3), activation='relu'),
            layers.BatchNormalization(),
            layers.MaxPooling2D((2, 2)),
            layers.Dropout(0.25),
            
            # Block 3
            layers.Conv2D(128, (3, 3), activation='relu'),
            layers.BatchNormalization(),
            layers.MaxPooling2D((2, 2)),
            layers.Dropout(0.25),
            
            # Dense layers
            layers.Flatten(),
            layers.Dense(256, activation='relu'),
            layers.BatchNormalization(),
            layers.Dropout(0.5),
            layers.Dense(128, activation='relu'),
            layers.Dropout(0.5),
            layers.Dense(self.num_classes, activation='softmax')
        ])
        
        model.compile(
            optimizer='adam',
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )
        
        self.model = model
        return model
    
    def train(self, X_train, y_train, X_val, y_val, epochs=30, batch_size=32):
        """Huấn luyện model"""
        
        callbacks = [
            keras.callbacks.EarlyStopping(
                monitor='val_loss',
                patience=10,
                restore_best_weights=True
            ),
            keras.callbacks.ReduceLROnPlateau(
                monitor='val_loss',
                factor=0.5,
                patience=5,
                min_lr=1e-7
            )
        ]
        
        self.history = self.model.fit(
            X_train, y_train,
            batch_size=batch_size,
            epochs=epochs,
            validation_data=(X_val, y_val),
            callbacks=callbacks,
            verbose=1
        )
        
        return self.history
    
    def evaluate(self, X_test, y_test):
        """Đánh giá model"""
        test_loss, test_acc = self.model.evaluate(X_test, y_test, verbose=0)
        print(f"Test Accuracy: {test_acc:.4f}")
        print(f"Test Loss: {test_loss:.4f}")
        return test_loss, test_acc
    
    def predict(self, image):
        """Dự đoán hình dạng từ ảnh"""
        if len(image.shape) == 2:
            image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
        
        img_resized = cv2.resize(image, self.input_shape[:2])
        img_normalized = img_resized.astype('float32') / 255.0
        img_batch = np.expand_dims(img_normalized, axis=0)
        
        predictions = self.model.predict(img_batch, verbose=0)
        class_idx = np.argmax(predictions[0])
        confidence = predictions[0][class_idx]
        
        return self.class_names[class_idx], confidence

# 3. VẼ BOUNDING BOX VÀ DETECTION
class BoundingBoxDetector:
    def __init__(self, cnn_model):
        self.cnn_model = cnn_model
        
    def detect_shapes(self, image):
        """Phát hiện hình dạng và trả về bounding box"""
        # Chuyển sang grayscale
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        
        # Áp dụng threshold
        _, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)
        
        # Tìm contours
        contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        
        detections = []
        
        for contour in contours:
            # Lọc contours nhỏ
            area = cv2.contourArea(contour)
            if area < 500:
                continue
                
            # Lấy bounding box
            x, y, w, h = cv2.boundingRect(contour)
            
            # Crop hình dạng
            shape_roi = image[y:y+h, x:x+w]
            
            if shape_roi.size == 0:
                continue
                
            # Phân loại bằng CNN
            predicted_shape, confidence = self.cnn_model.predict(shape_roi)
            
            detections.append({
                'bbox': (x, y, w, h),
                'shape': predicted_shape,
                'confidence': confidence,
                'contour': contour
            })
        
        return detections
    
    def draw_detections(self, image, detections):
        """Vẽ bounding boxes và labels lên ảnh"""
        result = image.copy()
        
        for detection in detections:
            x, y, w, h = detection['bbox']
            shape = detection['shape']
            confidence = detection['confidence']
            
            # Màu sắc dựa trên confidence
            if confidence > 0.9:
                color = (0, 255, 0)  # Xanh lá - độ tin cậy cao
            elif confidence > 0.7:
                color = (0, 165, 255)  # Cam - độ tin cậy trung bình
            else:
                color = (0, 0, 255)  # Đỏ - độ tin cậy thấp
            
            # Vẽ bounding box
            cv2.rectangle(result, (x, y), (x + w, y + h), color, 2)
            
            # Vẽ label với confidence
            label = f"{shape} ({confidence:.2f})"
            cv2.putText(result, label, (x, y - 10),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)
            
            # Vẽ contour
            cv2.drawContours(result, [detection['contour']], -1, (255, 0, 0), 1)
        
        return result

# 4. HIỂN THỊ KẾT QUẢ
def plot_training_history(history):
    """Vẽ biểu đồ quá trình huấn luyện"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
    
    # Biểu đồ accuracy
    ax1.plot(history.history['accuracy'], label='Training Accuracy')
    ax1.plot(history.history['val_accuracy'], label='Validation Accuracy')
    ax1.set_title('Model Accuracy')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Accuracy')
    ax1.legend()
    ax1.grid(True)
    
    # Biểu đồ loss
    ax2.plot(history.history['loss'], label='Training Loss')
    ax2.plot(history.history['val_loss'], label='Validation Loss')
    ax2.set_title('Model Loss')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Loss')
    ax2.legend()
    ax2.grid(True)
    
    plt.tight_layout()
    plt.show()

def plot_predictions(model, X_test, y_test, class_names, num_samples=12):
    """Hiển thị dự đoán trên test set"""
    # Lấy dự đoán
    y_pred = model.model.predict(X_test)
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_true_classes = np.argmax(y_test, axis=1)
    
    # Chọn ngẫu nhiên các sample
    indices = np.random.choice(len(X_test), min(num_samples, len(X_test)), replace=False)
    
    fig, axes = plt.subplots(3, 4, figsize=(15, 12))
    axes = axes.ravel()
    
    for i, idx in enumerate(indices):
        img = X_test[idx]
        true_class = class_names[y_true_classes[idx]]
        pred_class = class_names[y_pred_classes[idx]]
        confidence = y_pred[idx][y_pred_classes[idx]]
        
        # Hiển thị ảnh
        axes[i].imshow(img)
        axes[i].set_title(f'True: {true_class}\nPred: {pred_class}\nConf: {confidence:.2f}')
        
        # Đổi màu title nếu dự đoán sai
        if true_class != pred_class:
            axes[i].title.set_color('red')
        else:
            axes[i].title.set_color('green')
            
        axes[i].axis('off')
    
    plt.tight_layout()
    plt.show()
    
    # Confusion matrix
    cm = confusion_matrix(y_true_classes, y_pred_classes)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

# 5. DEMO HOÀN CHỈNH
def main():
    dataset_path = '/kaggle/input/geometric-shapes-dataset/geometric shapes dataset/'
    
    # Load data
    images, labels, class_names = load_and_prepare_data(dataset_path)
    
    if len(images) == 0:
        print("Không tìm thấy ảnh nào! Vui lòng kiểm tra đường dẫn dataset.")
        return
    
    # Chia data
    X_train, X_test, y_train, y_test = train_test_split(
        images, labels, test_size=0.2, random_state=42, stratify=np.argmax(labels, axis=1)
    )
    
    X_train, X_val, y_train, y_val = train_test_split(
        X_train, y_train, test_size=0.2, random_state=42
    )
    
    print(f"\n Data Split:")
    print(f"  Training: {len(X_train)} samples")
    print(f"  Validation: {len(X_val)} samples")
    print(f"  Test: {len(X_test)} samples")
    
    # Khởi tạo và huấn luyện model
    detector = ShapeDetectorCNN()
    detector.build_model()
    
    print("\n" + "="*50)
    print("MODEL ARCHITECTURE")
    print("="*50)
    detector.model.summary()
    
    # Huấn luyện
    history = detector.train(X_train, y_train, X_val, y_val, epochs=50)
    
    # Đánh giá
    detector.evaluate(X_test, y_test)
    
    # Vẽ biểu đồ
    plot_training_history(history)
    plot_predictions(detector, X_test, y_test, class_names)
    
    # 6. TEST VỚI ẢNH MỚI VÀ BOUNDING BOX
    print("\n" + "="*50)
    print("TEST VỚI BOUNDING BOX DETECTION")
    print("="*50)
    
    # Tạo ảnh test với nhiều hình dạng
    test_image = np.ones((300, 400, 3), dtype=np.uint8) * 255
    
    # Vẽ các hình dạng test
    cv2.circle(test_image, (80, 80), 30, (255, 0, 0), -1)  # Circle - Blue
    cv2.rectangle(test_image, (150, 50), (250, 150), (0, 255, 255), -1)  # Square 
    triangle_pts = np.array([[320, 50], [280, 150], [360, 150]], np.int32)
    cv2.fillPoly(test_image, [triangle_pts], (0, 0, 255))  # Triangle - Red
    
    # Khởi tạo bounding box detector
    bbox_detector = BoundingBoxDetector(detector)
    
    # Detect shapes
    detections = bbox_detector.detect_shapes(test_image)
    
    # Vẽ kết quả
    result_image = bbox_detector.draw_detections(test_image, detections)
    
    # Hiển thị kết quả
    plt.figure(figsize=(15, 5))
    
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(test_image, cv2.COLOR_BGR2RGB))
    plt.title('Ảnh Gốc')
    plt.axis('off')
    
    
    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))
    plt.title('Kết Quả Detection')
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()
    
    # In thông tin detection
    print("\n Detection Results:")
    for i, detection in enumerate(detections):
        print(f"  Shape {i+1}: {detection['shape']} (Confidence: {detection['confidence']:.2f})")
        print(f"    BBox: {detection['bbox']}")
    
    # Lưu model
    detector.model.save('shape_detector_3classes.h5')
    print(f"\n Model đã được lưu: shape_detector_3classes.h5")

# Chạy chương trình
if __name__ == "__main__":
    main()
```

## Nhận diện hình dạng đơn giản của đồ vật

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
import os

# ----------- FIX UNICODE ------------
def cv2_imread_unicode(path):
    """Đọc ảnh với đường dẫn Unicode (Windows)"""
    with open(path, "rb") as f:
        data = f.read()
    img_array = np.frombuffer(data, np.uint8)
    return cv2.imdecode(img_array, cv2.IMREAD_COLOR)

# 1. LOAD MODEL
def load_trained_model():
    model_path = r'H:\DOCUMENT\Xử lý ảnh\shape_detector_3classes.h5'
    
    if os.path.exists(model_path):
        model = keras.models.load_model(model_path)
        print("Đã load model CNN đã train")
        return model
    else:
        print("Không tìm thấy model!")
        return None

# 2. CHUYỂN ẢNH THÀNH ẢNH VIỀN
def convert_to_edge_image(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (5, 5), 1.5)
    edges = cv2.Canny(blurred, 50, 150)
    edge_rgb = cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR)
    return edge_rgb, edges

# 3. PHÂN LOẠI ẢNH VIỀN
def classify_edge_image_with_cnn(model, image):
    class_names = ['Circle', 'Square', 'Triangle']
    
    edge_rgb, edges_binary = convert_to_edge_image(image)
    
    img_resized = cv2.resize(edge_rgb, (128, 128))
    img_normalized = img_resized.astype('float32') / 255.0
    img_batch = np.expand_dims(img_normalized, axis=0)
    
    predictions = model.predict(img_batch, verbose=0)
    class_idx = np.argmax(predictions[0])
    confidence = predictions[0][class_idx]
    predicted_class = class_names[class_idx]
    
    return predicted_class, confidence, predictions[0], edge_rgb, edges_binary

# 4. HIỂN THỊ KẾT QUẢ
def analyze_edge_images(model, image_paths):
    num_images = len(image_paths)
    fig, axes = plt.subplots(num_images, 5, figsize=(20, 4 * num_images))
    
    if num_images == 1:
        axes = axes.reshape(1, -1)
    
    results = []
    
    for idx, image_path in enumerate(image_paths):
        if not os.path.exists(image_path):
            print(f"Không tìm thấy ảnh: {image_path}")
            continue
        
        # ========= FIX UNICODE HERE ==========
        image = cv2_imread_unicode(image_path)
        # =====================================
        
        if image is None:
            print(f"Không đọc được ảnh: {image_path}")
            continue
        
        print(f"Kích thước gốc: {image.shape[1]}x{image.shape[0]}")
        
        pred_class, confidence, probs, edge_rgb, edges_binary = classify_edge_image_with_cnn(model, image)
        
        results.append({
            'filename': os.path.basename(image_path),
            'prediction': pred_class,
            'confidence': confidence,
            'probs': probs
        })
        
        # Hiển thị
        axes[idx, 0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        axes[idx, 0].set_title(os.path.basename(image_path))
        axes[idx, 0].axis('off')
        
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        axes[idx, 1].imshow(gray, cmap='gray')
        axes[idx, 1].set_title('Grayscale')
        axes[idx, 1].axis('off')
        
        axes[idx, 2].imshow(edges_binary, cmap='gray')
        axes[idx, 2].set_title('Edges (Canny)')
        axes[idx, 2].axis('off')
        
        axes[idx, 3].imshow(cv2.cvtColor(edge_rgb, cv2.COLOR_BGR2RGB))
        axes[idx, 3].set_title('Edges RGB')
        axes[idx, 3].axis('off')
        
        axes[idx, 4].axis('off')
        axes[idx, 4].text(0.1, 0.8, f'{pred_class}', fontsize=12, fontweight='bold')
        axes[idx, 4].text(0.1, 0.7, f'Confidence: {confidence:.3f}', fontsize=10)
    
    plt.tight_layout()
    plt.show()
    
    return results



def main():
    print("=" * 60)
    model = load_trained_model()
    if model is None:
        return
    
    test_dir = r"H:\DOCUMENT\test"
    
    valid_images = [
        os.path.join(test_dir, f)
        for f in os.listdir(test_dir)
        if f.lower().endswith((".png", ".jpg", ".jpeg", ".bmp"))
    ]
    
    print(f"Đã tìm {len(valid_images)} ảnh để test.")
    
    results = analyze_edge_images(model, valid_images)
    

if __name__ == "__main__":
    main()
```
